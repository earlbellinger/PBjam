# The mixed mode asymptotic relation

The Mosser mixed mode relation is given as:

$$ \tan \frac{\pi(\nu - \nu_{\rm p})}{\Delta \nu} = q \, \tan \left(\frac{\pi}{\Delta \Pi \, \nu} - \epsilon_{\rm g}\right), $$

where the roots of the equation are the mixed mode frequencies.

We can approximate $\nu_{\rm p}$ with the p=mode asymptotic relation:

$$ \nu_{\rm p} \approx (n + \epsilon + d_{01}) \Delta \nu ,$$

where we could use higher order terms if required.

We want to replace $\nu - \mu_{\rm p}$ with a new function expressed with $\Delta \nu$ to remove the dependence on $n$ and the multiple solutions.  

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = np.arange(10, 11, 1)
dnu = 16.97
epsilon = 1.4
d01 = 0.5
nu_p = (n + epsilon + d01) * dnu
nu_p
```

```python
nu = np.linspace(115, 280, 10000)
q = 0.14
dp1 = 70.0
epsilon_g = 0.0
lhs = [np.tan(np.pi * (nu - p) / dnu) for p in nu_p]
rhs = q * np.tan(np.pi / dp1 / nu*1e6 - epsilon_g) 
```

```python
fig, ax = plt.subplots(figsize=[16,9])
for idx, lh in enumerate(lhs):
    ax.plot(nu, rhs - lh + idx)
ax.set_ylim([-10, 10])
```
