```python
import numpy as np
import cpnest.model
```

```python
class Simple2DModel(cpnest.model.Model):
    """
    A simple 2 dimensional gaussian
    """
    names=['x','y']
    bounds=[[-10,10],[-10,10]]

    def log_likelihood(self, param):
        return -0.5*(param['x']**2 + param['y']**2) - np.log(2.0*np.pi)
```

```python
mymodel = Simple2DModel()
nest = cpnest.CPNest(mymodel, nlive=1024)
nest.run()
```

```python
import pandas as pd
df = pd.DataFrame(nest.get_posterior_samples())
df.head()
```

```python
import corner 
corner.corner(df[['x', 'y', 'logL']]);
```
